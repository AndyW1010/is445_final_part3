```python
import pandas as pd
import matplotlib.pyplot as plt
import altair as alt
import ipywidgets as widgets
df = pd.read_excel('is445.xlsx')
df = df.drop(df.index[60:62])
df.replace('NR', float('0'), inplace=True)
df['Total'] = df.select_dtypes(include='number').sum(axis=1)
df
```

The code provided generates a bar plot that displays the top 10 diseases with the highest number of cases in Illinois from 2010 to 2017. The data is first sorted in descending order based on the total number of cases, and the top 10 diseases are selected. The resulting bar plot shows the name of each disease on the x-axis and the corresponding total number of cases on the y-axis. The plot aims to highlight the most prevalent diseases in Illinois during the specified time period, which can be helpful for healthcare professionals and policymakers in allocating resources and designing public health interventions.I think it's interesting to study this because it can help us understand the most prevalent diseases so that we can better deal with them

```python
df.columns = df.columns.astype(str)
chart = alt.Chart(df).mark_bar().encode(
    x='Disease',
    y='Total'
)
chart
```

```python
bar_chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('Disease', axis=alt.Axis(titleFontSize=16)),
    y=alt.Y('Total', axis=alt.Axis(titleFontSize=16)),
    color=alt.condition(
        alt.selection_single(name='highlight', empty='none', on='mouseover'),
        alt.value('steelblue'),
        alt.value('grey')  
    ),
    tooltip=['Disease', 'Total']
).properties(
    selection=alt.selection_single(name='highlight', empty='none', on='mouseover')
)
line_chart = alt.Chart(df).mark_line().encode(
    x=alt.X('Year:quantitative', axis=alt.Axis(format='d', titleFontSize=16)),
    y=alt.Y('value:quantitative', axis=alt.Axis(title='Cases', titleFontSize=16)),
    color=alt.Color('Disease', scale=alt.Scale(range=['#1f77b4']))
).transform_fold(
    fold=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'],
    as_=['Year', 'value']
).properties(
    width=400,
    height=300,
    title=alt.TitleParams(text='Number of Cases by Year', fontSize=20) 
)

selection = alt.selection_single(empty='none', on='mouseover', fields=['Disease'])
dashboard = (bar_chart.add_selection(selection) & line_chart.transform_filter(selection)).properties(
    title={
        "text": "Interactive Dashboard",
        "subtitle": ["Exploring Disease Cases by Year", "and Disease Type"],
        "color": "black",
        "subtitleColor": "gray",
        "fontSize": 26,
        "subtitleFontSize": 18,
        "align": "center",
        "anchor": "middle",
        "fontWeight": "bold", 
        "offset": 20 
    },
    padding={"left": 50, "top": 50, "right": 50, "bottom": 50},
    background='white'
)

dashboard
```

The interactive dashboard I have created allows non-experts to easily explore and visualize the number of reported cases for different diseases over time. The bar chart displays the total number of cases for each disease, with the option to highlight a specific disease by hovering over it. In the barchart, the highlighted bar is colored in steel blue and the non-highlighted bars are grey.The line chart shows the number of cases per year, with the option to filter by disease type. By selecting a disease in the bar chart, the line chart will update to display only that disease's cases over time. This provides an easy way to see how the number of cases has changed for a specific disease compared to others in the dataset. Additionally, hovering over a bar displays the disease name and total number of cases, providing users with more detailed information.

```python
df1 = pd.read_csv("odp_idb_2021_ddg_compliant.csv")
df1
```

```python
Disease_cases = df1.groupby("Disease").sum()
Disease_cases = Disease_cases.reset_index()
Disease_cases = Disease_cases[Disease_cases["Cases"] > 1000]
Disease_cases[['Disease',"Cases"]]
```

```python
disease_cases_df = pd.DataFrame({'Disease': Disease_cases['Disease'], 'Cases': Disease_cases['Cases']})
cmap = alt.Scale(domain=(disease_cases_df['Cases'].min(), disease_cases_df['Cases'].max()), range=['#f7fbff', '#08306b'], type='linear')
chart = alt.Chart(disease_cases_df, title='Number of Reported Cases by Disease').mark_bar().encode(
    alt.X('Disease:O', axis=alt.Axis(labelAngle=-90, labelFontSize=16, title='Name of Disease')),
    alt.Y('Cases:Q', axis=alt.Axis(labelFontSize=16, title='Number of Cases')),
    color=alt.Color('Cases:Q', scale=cmap)
).configure_axis(
    labelFontSize=14,
    titleFontSize=16
).configure_view(
    height=400,
    width=800
).configure_title(
    fontSize=20,
    fontWeight='bold'
)
chart

```
